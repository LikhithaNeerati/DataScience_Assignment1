```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Data

df = pd.read_csv("C:/Users/Likhita/OneDrive/Documents/PDS/FrailtyData.csv")

# Data Cleaning and Exploration
# Displaying basic information about the dataset
print(df.info())
```

```python
ax=sns.distplot(df['Age'], kde = False, color ='blue', bins = 30)
fig=ax.get_figure()
fig.savefig("C:/Users/Likhita/OneDrive/Documents/Age_Distibution.png")
```

```python
ax=sns.distplot(df['Grip Strength'], kde = False, color ='green', bins = 30)
fig=ax.get_figure()
fig.savefig("C:/Users/Likhita/OneDrive/Documents/Grip_Strength_Distibution.png")
```

```python

# Handling missing values by removing rows with any missing data
df_cleaned = df.dropna()

# Summarizing the data
print(df_cleaned.describe())

# Visualization
# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Creating a scatter plot to explore the relationship between weight and grip strength
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='Weight', y='Grip Strength', hue='Age')
plt.title('Relationship between Weight and Grip Strength')
plt.xlabel('Weight (kg)')
plt.ylabel('Grip Strength (kg)')
plt.legend(title='Age')
plt.savefig('GripStrength_Weight_Relation.png')
plt.show()
```
